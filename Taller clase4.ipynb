```python
from pylab import plot, show

y = [1., 2.4, 1.7, 0.3, 0.6, 1.8]

plot(y)
show()
    
```

```python
from pylab import savefig
x = [0.5, 1., 2., 4., 7., 10.]
plot(x,y)
savefig('grafica_1.eps', format='eps', dpi=1000)
show()
```

```python
from numpy import linspace, sin ,cos
x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid
x = linspace(0, 10, 1000)
y = sin(x)
plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
show()
```

```python
from pylab import*
plot(x,y, 'g--')
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje x")
ylabel("Eje y")
show()
```

```python
from pylab import*
plot(x,y, 'g--')
y = cos(x)
plot(x,y, 'r-')
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje x")
ylabel("Eje y")
show()
```
